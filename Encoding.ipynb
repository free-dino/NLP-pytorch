# TF representation:
TF, which stands for term frequency, is like the sum of one-hot encoding

```python
def vocab_from_corpus(corpus):
    vocab = set()
    for sentence in corpus:
        for word in sentence.split():
            vocab.add(word.lower().strip('.,'))
    return sorted(vocab)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns
corpus = ['Time flies flies like an arrow.',
'Fruit flies like a banana.']
vocab = vocab_from_corpus(corpus)
one_hot_vectorizer = CountVectorizer(binary=True)
one_hot = one_hot_vectorizer.fit_transform(corpus).toarray()
sns.heatmap(one_hot, annot=True,
    cmap = 'Blues', xticklabels=vocab,
    yticklabels=['Sentence 1','Sentence 2'])
```

# TF-IDF Representation
TF-IDF stands for term frequency - inverse document frequency.

$$
\text{IDF}(w) = \log \frac{N}{n_w}
$$

where $n_w$ is the number of documents containing the word $w$ and $N$ is the total number of documents

$$
\text{TF-IDF}(w) = \text{TF}(w) \cdot \text{IDF}(w)
$$

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf = tfidf_vectorizer.fit_transform(corpus).toarray()
sns.heatmap(tfidf, annot=True, cmap='Blues', xticklabels=vocab,
            yticklabels= ['Sentence 1', 'Sentence 2'])
```
